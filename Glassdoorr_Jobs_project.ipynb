## 3. ***Data Wrangling***

### Data Wrangling Code

````python
df = df[df['Salary Estimate'] != '-1']
````

````python

# Extract salary range and convert to numeric values
df['Salary Estimate'] = df['Salary Estimate'].str.replace(r'\(Glassdoor est.\)', '', regex=True).str.strip()
df[['Min Salary', 'Max Salary']] = df['Salary Estimate'].str.extract(r'\$(\d+)K?-\$(\d+)K?')
df['Min Salary'] = pd.to_numeric(df['Min Salary']) * 1000
df['Max Salary'] = pd.to_numeric(df['Max Salary']) * 1000
df['Avg Salary'] = (df['Min Salary'] + df['Max Salary']) / 2
````

````python
# Clean company names by removing ratings
df['Company Name'] = df['Company Name'].str.replace(r'\n\d+\.\d+', '', regex=True).str.strip()
````

````python
# Extract city and state from location
df[['City', 'State']] = df['Location'].str.rsplit(',', n=1, expand=True)
df['City'] = df['City'].str.strip()
df['State'] = df['State'].str.strip()
````

````python
import numpy as np

def clean_revenue(value):
    value = str(value).lower().strip()  # Ensure it's a string and lowercase

    if 'million' in value:
        num = value.split()[0].replace('$', '').replace('+', '')  # Remove "$" and "+"
        if num.isdigit():  # Check if it's a valid number
            return int(num) * 1e6
    elif 'billion' in value:
        num = value.split()[0].replace('$', '').replace('+', '')  # Remove "$" and "+"
        try:
            return float(num) * 1e9  # Convert to float and scale
        except ValueError:
            return np.nan  # If conversion fails, return NaN

    return np.nan  # Default to NaN for unexpected values

df['Revenue'] = df['Revenue'].apply(clean_revenue)

````

````python
# Replace '-1' in competitors with NaN
df['Competitors'] = df['Competitors'].replace('-1', np.nan)
````

````python
df['Sector'] = df['Sector'].replace('-1', np.nan)
````

````python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Combine all job descriptions into a single string
text = ' '.join(df['Job Description'].astype(str).tolist())

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=STOPWORDS).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Job Descriptions')
plt.show()
````
